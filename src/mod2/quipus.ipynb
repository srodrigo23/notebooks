```python
import random
from faker import Faker
import pandas as pd
# from datetime import datetime
# import csv

fake = Faker('es_ES')

# Crear instancia de Faker
# Configuracin
# num_equipos = 10
num_ventas = 10000

# Posibles valores para algunos campos
# nombres_equipos = ['computadora de escritorio', 'laptop']
# nombres_equipos_pesos = [0.2, 0.8]

# class NombresEquiposProvider:
#     def nombreEquipo(self):
#         return random.choices(nombres_equipos, weights=nombres_equipos_pesos, k=1)[0]
# fake.add_provider(NombresEquiposProvider)

ciudades = ['Oruro', 'La Paz', 'Potosi', 'Santa Cruz', 'Cochabamba', 'Tarija', 'Chuquisaca', 'Beni', 'Pando']
ciudades_pesos = [0.03, 0.17, 0.1, 0.13, 0.06, 0.15, 0.12, 0.06, 0.18]

class CiudadesProvider:
    def nombreCiudad(self):
        return random.choices(ciudades, weights=ciudades_pesos, k=1)[0]
fake.add_provider(CiudadesProvider())

tipo_clientes = ['Persona', 'Institucin']
tipo_clientes_pesos = [0.3, 0.7]

class TipoClienteProvider:
    def tipoCliente(self):
        return random.choices(tipo_clientes, weights=tipo_clientes_pesos, k=1)[0]
fake.add_provider(TipoClienteProvider())

tipo_instituciones = ['Pblica', 'Privada']
tipo_instituciones_pesos = [0.8, 0.2]

class TipoInstitucionProvider:
    def tipoInstitucion(self):
        return random.choices(tipo_instituciones, weights=tipo_instituciones_pesos, k=1)[0]
fake.add_provider(TipoInstitucionProvider())

tipo_adquisiciones = ['Compra directa', 'Licitacin']
tipo_adquisiciones_pesos = [0.4, 0.6]

class TipoAdquisicionesProvider:
    def tipoAdquisiciones(self):
        return random.choices(tipo_adquisiciones, weights=tipo_adquisiciones_pesos, k=1)[0]
fake.add_provider(TipoAdquisicionesProvider())

tipo_compras = ['Tienda', 'en Lnea']
tipo_compras_pesos = [0.85, 0.35]

class TipoComprasProvider:
    def tipoCompras(self):
        return random.choices(tipo_compras, weights=tipo_compras_pesos, k=1)[0]
fake.add_provider(TipoComprasProvider())

nombre_institucion=[
    'Gobierno Municipal de Trinidad',
    'Gobierno Autnomo Departamental de Pando',
    'Gobierno Autnomo Municipal de Cobija',
    'Gobierno Autnomo Departamental de Chuquisaca',
    'Gobierno Municipal de La Paz',
    'Gobierno Autnomo Departamental de La Paz',
    'Gobierno Autnomo Departamental de Beni',
    'Gobierno Autnomo Departamental de Tarija',
    'Gobierno Autnomo Municipal de Tarija',
    'Gobierno Autnomo Departamental de Potos',
    'Gobierno Autnomo Departamental de Oruro',
    'Gobierno Autnomo Departamental de Cochabamba',
    'Gobierno Autnomo Departamental de Santa Cruz',
    'Concejo Municipal de Santa Cruz'
]

df_equipos= pd.read_csv("dataset/equipos.csv", encoding='utf-8')

# Generar datos para la tabla Venta
ventas = []
for i in range(num_ventas):

    # Faker.seed(random.choice(range(100))) #this approach don't works
    fecha_pedido = fake.date_between(start_date='-5y', end_date='today')
    fecha_entrega = fake.date_between(start_date=fecha_pedido, end_date='+30d')

    tipo_cliente = fake.tipoCliente() #random.choice(tipo_clientes)
    tipo_institucion = fake.tipoInstitucion() if tipo_cliente == 'Institucin' else 'NaN' #random.choice(tipo_instituciones) 
    
    if (tipo_cliente == 'Institucin') and (tipo_institucion == 'Pblica'):
        cliente = random.choice(nombre_institucion) #??????
    elif (tipo_cliente == 'Institucin') and (tipo_institucion == 'Privada'):
        cliente = fake.company()
    else: cliente = fake.name()

    ciudad = fake.nombreCiudad()#random.choice(ciudades) #???????
    direccion = fake.address()
    tipo_adquisicion = fake.tipoAdquisiciones()#random.choice(tipo_adquisiciones)
    tipo_compra = fake.tipoCompras()#random.choice(tipo_compras)
    cantidad = random.randint(1, 100)
    cod_equipo = random.choices(
        df_equipos['cod_equipo'], 
        (0.08, 0.06, 0.03, 0.09, 0.08, 0.02, 0.05, 0.06, 0.07, 0.04, 0.05, 0.12, 0.05, 0.09, 0.1)
    )[0]
    print(cod_equipo)
    precio_unitario = df_equipos.loc[df_equipos['cod_equipo'] == cod_equipo, 'precio_unitario'].values[0]
    monto_total = cantidad * precio_unitario

    ventas.append({
        'idVenta': fake.unique.random_number(digits=6),
        'FechaPedido': fecha_pedido,
        'FechaEntrega': fecha_entrega,
        'TipoCliente': tipo_cliente,
        'TipoInstitucion': tipo_institucion,
        'Cliente': cliente,
        'Ciudad': ciudad,
        'Direccion': direccion,
        'TipoAdquisicion': tipo_adquisicion,
        'TipoCompra': tipo_compra,
        'precio_unitario': precio_unitario,
        'Cantidad': cantidad,
        'CodEquipo': cod_equipo,
        'MontoTotal': monto_total
    })

# Convertir a DataFrame
df_ventas = pd.DataFrame(ventas)

# Mostrar los primeros registros de cada DataFrame
#print(df_equipos.head())
#print(df_ventas.head())

# Opcional: guardar los datos en archivos CSV
# df_equipos.to_csv('equipos.csv', index=False)
df_ventas.to_csv('ventas.csv', index=False)
```

```python
def find_nine_values_sum_to_100():
  values = [random.randint(1, 15) for _ in range(14)]  # Generate 8 random values
  remaining_sum = 100 - sum(values)
  values.append(remaining_sum)  # Add the ninth value to ensure the total is 100
  return values


result = find_nine_values_sum_to_100()
result = list((map(lambda n: n/100,result)))

print(result)

print("Sum:", sum(result))
```
